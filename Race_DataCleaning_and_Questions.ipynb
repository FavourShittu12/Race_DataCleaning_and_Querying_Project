```python
#Extracting races in the USA, with 50km or 50 miles  IN 2020
```

```python

new_df = df[((df['Event distance/length'] == '50mi') | (df['Event distance/length'] == '50km')) & (df['Year of event'] == 2020) & (df['Athlete country'] == 'USA') &( df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
new_df
```

```python
new_df.shape
```

```python
#Removing "(USA)" from event name
```

```python
new_df['Event name'].str.split('(').str.get(0)
```

```python
new_df['Event name'] = new_df['Event name'].str.split('(').str.get(0)
```

```python
#Cleaning up athlete age
new_df['Athlete age'] =  (new_df['Year of event'] - new_df['Athlete year of birth']).fillna(0).astype(int)
```

```python
new_df.head(5)
```

```python
# Cleaning up the Athlete performance column: Removing h
new_df['Athlete performance'] = new_df['Athlete performance'].str.split(' ').str.get(0)
```

```python
#Cleaning up the Year of Birth Column - Removing decimal points
new_df['Athlete year of birth'] = new_df['Athlete year of birth'].fillna(0).astype(int)
new_df
```

```python
# drop Columns : Athlete Club, Athlete Country
#Club and Country have been dropped  
#new_df.drop('Athlete club', axis = 1, inplace = True)
#new_df.drop('Athlete country', axis = 1, inplace = True)
new_df.head(5)
```

```python
#Cleani up the remaining missing rows
new_df.isna().sum()
```

```python
new_df = new_df.dropna()
```

```python
new_df = new_df.drop('Year of event', axis = 1)
```

```python
#Questions
```

```python
#1.)Difference in speed for the 50k and 50mi male to female
new_df.groupby(['Event distance/length','Athlete gender'])['Athlete average speed'].mean()
```

```python
#2.)What age groups are the best in the 50mi Race minimum of 10 races

new_df.query('`Event distance/length` == "50mi"').groupby('Athlete age')['Athlete average speed'].agg(['mean','count']).sort_values(by='mean',ascending=False).query('count >= 10')
```

```python
#3.)What age groups are the worst in the 50mi Race minimum of 10 races

new_df.query('`Event distance/length` == "50mi"').groupby('Athlete age')['Athlete average speed'].agg(['mean','count']).sort_values(by='mean',ascending=True).query('count >= 10')
```
